```python
import pandas as pd
```

```python
df=pd.read_csv("Cleaned_Data.csv")
```

```python
df
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
train_data = df[0:150000]
train_label = train_data["price"]
train_data.drop("price", inplace= True, axis = 1);
```

```python
train_data
```

```python
train_label
```

```python
test_data = df[150000:]
test_label = test_data["price"]
test_data.drop("price", inplace= True, axis = 1)
```

```python
test_label
```

```python
model = MLPRegressor(activation = "relu", learning_rate_init = 0.01, hidden_layer_sizes =[50])
```

```python
model.fit(train_data, train_label)
```

```python
predictions = model.predict(test_data)
```

```python
predictions
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(test_label, predictions)
```

```python
from sklearn.model_selection import cross_validate
```

Cross Validation of the model

```python
data_set = df
data_label = data_set["price"]
data_set.drop("price", axis = 1, inplace = True)
```

```python
s = cross_validate(model, data_set,data_label, cv =3, return_train_score= True)
```

```python
s
```

R2 Score after cross validation

```python
s["test_score"].mean()
```

Final Model training

```python
model = MLPRegressor(activation = "relu", learning_rate_init = 0.01, hidden_layer_sizes =[50], max_iter = 1000)
```

```python
model.fit(data_set, data_label)
```
